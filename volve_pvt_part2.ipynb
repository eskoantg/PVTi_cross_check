```python
# =============================================================================
# Cell 2.2d — Back-calculate what A, B PVTi must be using
# =============================================================================
from scipy.optimize import fsolve
import numpy as np

Z_pvti  = 1.0793
P_bp    = 200.8

print('=== Cell 2.2d — Back-calculate PVTi A, B ===\n')
print(f'  PVTi Z = {Z_pvti}  at P = {P_bp} bar\n')

# If Z=1.0793 is the ONLY real root, the cubic must have the form:
# (Z - Z_pvti)(Z^2 + pZ + q) = 0  where Z^2+pZ+q has no real roots
# Expanding: Z^3 + (p - Z_pvti)Z^2 + (q - p*Z_pvti)Z - q*Z_pvti = 0
# 
# Matching with PR cubic:
# Z^3 + (B-1)Z^2 + (A-2B-3B^2)Z + (B^3+B^2-AB) = 0
# 
# From Z coefficient:  B - 1 = p - Z_pvti  => p = B - 1 + Z_pvti
# From constant term:  -q*Z_pvti = B^3+B^2-AB
# 
# We have two equations and two unknowns (A, B).
# Use PVTi's reported molar volume to get another constraint:
# V_pvti = Z_pvti * R * T / P  (raw PR volume before shift)
# OR
# V_pvti = 169.8 cm3/mol (reported) = Z_pvti_raw * R * T / P

# If 169.8 is the SHIFTED volume:
# V_shifted = V_raw - C_MIX  => V_raw = V_shifted + C_MIX
V_shifted = 169.8
V_raw     = V_shifted + C_MIX
Z_raw     = V_raw * P_bp / (R * T)

print(f'  If V_pvti=169.8 is shifted volume:')
print(f'    V_raw = V_shifted + C_MIX = {V_shifted} + {C_MIX:.3f} = {V_raw:.3f} cm3/mol')
print(f'    Z_raw = {Z_raw:.6f}')
print()

# If 169.8 is the RAW volume:
Z_from_raw = V_shifted * P_bp / (R * T)
print(f'  If V_pvti=169.8 is raw PR volume:')
print(f'    Z_raw = {Z_from_raw:.6f}')
print()

# Check: does PVTi Z=1.0793 correspond to raw or shifted?
V_check = Z_pvti * R * T / P_bp
print(f'  Z=1.0793 implies V = {V_check:.3f} cm3/mol')
print(f'  PVTi reports V_L  = 169.800 cm3/mol')
print(f'  Match: {"YES" if abs(V_check - V_shifted) < 0.5 else "NO"}')
print()

# So Z_pvti=1.0793 IS consistent with V=169.8 cm3/mol
# This means PVTi reports the SHIFTED Z (not raw PR Z)
# Raw Z = V_raw * P / (RT) where V_raw = V_shifted + C_MIX
Z_pvti_raw = (V_shifted + C_MIX) * P_bp / (R * T)
print(f'  PVTi shifted Z (reported)  = {Z_pvti:.4f}')
print(f'  PVTi raw PR Z (back-calc)  = {Z_pvti_raw:.6f}')
print(f'  Our calculated Z_L         = 0.963289')
print(f'  Difference raw             = {Z_pvti_raw - 0.963289:+.6f}')

# Now test cubic with Z_pvti_raw
print()
for B_test, label in [(B_raw_bp, 'B_raw'), (B_eff_bp, 'B_eff')]:
    A_test = A_bp
    f = (Z_pvti_raw**3
         + (B_test-1)*Z_pvti_raw**2
         + (A_test - 2*B_test - 3*B_test**2)*Z_pvti_raw
         + (B_test**3 + B_test**2 - A_test*B_test))
    print(f'  f(Z_raw={Z_pvti_raw:.4f}) with {label:7s} = {f:.8f}')
```

```python
# =============================================================================
# Cell 2.4 — Fugacity Coefficients
# =============================================================================
# 
# Uses Z_raw and B_raw in standard PR fugacity formula.
# Volume shift does NOT appear in fugacity (Peneloux 1982).
# 
# ln(phi_i) = (b_i/B_MIX)*(Z-1) - ln(Z-B)
# - A/(2√2·B) * (da_i/A_MIX - b_i/B_MIX)
# * ln[(Z+(1+√2)·B) / (Z+(1-√2)·B)]
# 
# where:
# Z   = Z_L_raw  (unshifted)
# B   = B_RAW    (unshifted)
# A   = A_DIM    (unshifted)
# da_i = 2 * sum_j(z_j * a_ij)
# =============================================================================

Z  = Z_L_raw
B  = B_RAW
A  = A_DIM
SQ2 = np.sqrt(2.0)

# section
DA_I = 2.0 * (A_IJ @ ZI)

# section
LOG_ARG = (Z + (1.0 + SQ2)*B) / (Z + (1.0 - SQ2)*B)

LN_PHI_L = (
      (B_I / B_MIX) * (Z - 1.0)
    - np.log(Z - B)
    - (A / (2.0 * SQ2 * B))
    * (DA_I / A_MIX - B_I / B_MIX)
    * np.log(LOG_ARG)
)

PHI_L = np.exp(LN_PHI_L)

# section
print('=== Cell 2.4 — Fugacity Coefficients (liquid phase at 213 bar) ===\n')
print(f'  Z_raw  = {Z:.6f}  (unshifted, used in fugacity)')
print(f'  B_raw  = {B:.6f}  (unshifted)')
print(f'  A      = {A:.6f}')
print()
print(f'  Z - B        = {Z - B:.6f}  (must be > 0)')
print(f'  log argument = {LOG_ARG:.6f}  (must be > 1)')
print()

df_24 = pd.DataFrame({
    'da_i'     : np.round(DA_I,    4),
    'ln_phi_i' : np.round(LN_PHI_L, 5),
    'phi_i'    : np.round(PHI_L,    6),
}, index=COMPONENTS)

display(df_24)

print()
print('Sanity checks:')
print(f'  N2  phi > 1  (volatile)  : {"✓" if PHI_L[0]  > 1 else "✗"}  phi={PHI_L[0]:.4f}')
print(f'  C1  phi > 1  (volatile)  : {"✓" if PHI_L[2]  > 1 else "✗"}  phi={PHI_L[2]:.4f}')
print(f'  C16+ phi << 1 (non-vol)  : {"✓" if PHI_L[11] < 0.01 else "✗"}  phi={PHI_L[11]:.6f}')
```

```python
# =============================================================================
# Cell 2.5 — Bubble Point Pressure Solver
# =============================================================================
from scipy.optimize import brentq

def get_AB(z, P_in):
    """calc A, B_raw, Z_raw for a given composition and pressure."""
    a_ij_loc = np.outer(np.sqrt(A_I), np.sqrt(A_I)) * (1.0 - K_IJ)
    a_mix    = float(np.einsum('i,ij,j', z, a_ij_loc, z))
    b_mix    = float(np.dot(z, B_I))
    A_loc    = a_mix  * P_in / (R * T)**2
    B_loc    = b_mix  * P_in / (R * T)
    return A_loc, B_loc, a_mix, b_mix, a_ij_loc

def solve_Z_raw(A_loc, B_loc, phase):
    """Solve PR cubic with B_raw. Returns Z_raw."""
    coeffs = [1.0,
              (B_loc - 1.0),
              (A_loc - 2.0*B_loc - 3.0*B_loc**2),
              (B_loc**3 + B_loc**2 - A_loc*B_loc)]
    roots     = np.roots(coeffs)
    real_r    = roots[np.abs(roots.imag) < 1e-8].real
    valid_r   = np.sort(real_r[real_r > B_loc])
    if len(valid_r) == 0:
        return B_loc + 1e-8
    elif len(valid_r) == 1:
        return valid_r[0]
    else:
        return valid_r[0] if phase == 'L' else valid_r[-1]

def ln_phi_calc(z, P_in, phase):
    """Returns ln(phi_i) using Z_raw and B_raw."""
    A_loc, B_loc, a_mix, b_mix, a_ij_loc = get_AB(z, P_in)
    Z   = solve_Z_raw(A_loc, B_loc, phase)
    SQ2 = np.sqrt(2.0)
    da_i = 2.0 * (a_ij_loc @ z)
    lnphi = (
          (B_I / b_mix) * (Z - 1.0)
        - np.log(Z - B_loc)
        - (A_loc / (2.0 * SQ2 * B_loc))
        * (da_i / a_mix - B_I / b_mix)
        * np.log((Z + (1.0+SQ2)*B_loc) / (Z + (1.0-SQ2)*B_loc))
    )
    return lnphi, Z

def bubble_point_residual(P_in, z):
    """
    At bubble point: liquid = z, sum(K_i*z_i) = 1
    K_i = exp(lnphi_L_i - lnphi_V_i)
    Successive substitution — 200 iterations max.
    """
    # Wilson K initial guess
    K = (PC / P_in) * np.exp(5.373*(1.0 + OMEGA)*(1.0 - TC/T))

    for _ in range(200):
        y            = K * z
        y            = y / y.sum()
        lnphi_L, _   = ln_phi_calc(z, P_in, 'L')
        lnphi_V, _   = ln_phi_calc(y, P_in, 'V')
        K_new        = np.exp(lnphi_L - lnphi_V)
        err          = np.sum((np.log(K_new/K))**2)
        K            = K_new
        if err < 1e-12:
            break

    return float(np.sum(K * z) - 1.0)

# section
print('=== Cell 2.5 — Bubble Point Pressure Solver ===\n')
print('Scanning residual vs pressure:')

P_scan = np.arange(50, 260, 10)
res    = [bubble_point_residual(float(p), ZI) for p in P_scan]

for p, r in zip(P_scan, res):
    print(f'  P = {p:5.0f} bar   residual = {r:+.5f}')

# section
P_lo, P_hi = None, None
for k in range(len(res)-1):
    if res[k] * res[k+1] < 0:
        P_lo = float(P_scan[k])
        P_hi = float(P_scan[k+1])
        break

print()
if P_lo is not None:
    P_bub = brentq(bubble_point_residual, P_lo, P_hi,
                   args=(ZI,), xtol=0.01)

    lnphi_L, Z_L_bp = ln_phi_calc(ZI, P_bub, 'L')
    K_fin = (PC/P_bub)*np.exp(5.373*(1+OMEGA)*(1-TC/T))
    for _ in range(200):
        y = K_fin*ZI; y /= y.sum()
        lnphi_V, Z_V_bp = ln_phi_calc(y, P_bub, 'V')
        K_new = np.exp(lnphi_L - lnphi_V)
        if np.sum((np.log(K_new/K_fin))**2) < 1e-12: break
        K_fin = K_new
    y_bp = K_fin*ZI; y_bp /= y_bp.sum()

    # Shifted Z at bubble point
    C_MIX_bp = float(np.dot(ZI, C_I))
    Z_L_shift = Z_L_bp - C_MIX_bp * P_bub / (R*T)

    print(f'  ✓  Bubble point        = {P_bub:.2f} bar')
    print(f'     PVTi target         = 200.84 bar')
    print(f'     Δ                   = {P_bub-200.84:+.2f} bar')
    print()
    print(f'     Z_L raw             = {Z_L_bp:.5f}')
    print(f'     Z_L shifted         = {Z_L_shift:.5f}  (PVTi: 1.0793)')
    print(f'     Z_V raw             = {Z_V_bp:.5f}')
    print()

    df_kv = pd.DataFrame({
        'z_i'  : np.round(ZI,    5),
        'y_i'  : np.round(y_bp,  5),
        'K_i'  : np.round(K_fin, 5),
    }, index=COMPONENTS)
    print('K-values and incipient vapour:')
    display(df_kv)
else:
    print('  ✗  No sign change found in 50-250 bar.')
    print('     Check residual scan above.')
```
